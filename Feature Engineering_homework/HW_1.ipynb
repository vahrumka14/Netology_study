# Категорийные атрибуты

а. Посчитать среднее кол-во комнат в строке (total_rooms/households)

```python
df["rooms_mean"] = df["total_rooms"]/df["households"]
```

```python
df.head()
```

```python
df['int_rooms_mean'] = df['rooms_mean'].apply(round)
```

```python
df.head()
```

б. Разбить полученный столбец на группы на ваше усмотрение (Например, 1 комната, 2 комнаты, 3 и более комнат)

```python
def classif(row):
    if row['int_rooms_mean'] == 1:
        return '1 комната'
    elif row['int_rooms_mean'] == 2:
        return '2 комнаты'
    elif row['int_rooms_mean'] == 3:
        return '3 комнаты'
    elif row['int_rooms_mean'] == 4:
        return '4 комнаты'
    elif row['int_rooms_mean'] >= 5:
        return 'больше 5 комнат'
    else:
        return 'undefined'
```

```python
df['class'] = df.apply(classif, axis=1)
```

```python
df.groupby('class').count()[['int_rooms_mean']].sort_values('int_rooms_mean', ascending=False)
```
