```python
# Ручной алгоритм LogisticRegression
```

```python
# шаг алгоритма
alpha = 0.01
```

```python
def gradient_descent_logisticregression(alpha, x, y, numIterations):
    """
    Функция реализует алгоритм градиентного спуска. На каждом шаге выводится значение функции потерь
    """
    
    m = x.shape[0] # 100
    theta = np.ones(5) # произвольно задаем начальные коэффициенты
    x_transpose = x.transpose() # транспонированная матрица x
    
    for iter in range( 0, numIterations ):
        matr = x@theta # матричное произведение
        
        hypothesis = 1/(1+np.exp(-matr)) # наш прогнозный таргет
        
        logloss = -y*np.log(hypothesis) - (1-y)*np.log(1-hypothesis)      
        
        J = np.sum(logloss)/(m) 

        print("J : ", J)
        
        gradient = (hypothesis-y)@x     
        
        theta = theta - alpha * gradient  
    
    return theta
```

```python
def predict_y(x,theta):
    matr = x@theta
    hypothesis = 1/(1+np.exp(-matr))
    return hypothesis
```

```python
%%time
theta = gradient_descent_logisticregression(alpha, x, y, 2000)
```

```python
print(theta)
```

```python
y_hat = predict_y(x, np.array(theta))
```

```python
def get_labels(y_hat, treshhold=0.5):
    y_corr = []
    for i in y_hat:
        if i >= treshhold:
            y_corr.append(1)
        else:
            y_corr.append(0)
    return y_corr
```

```python
y_corr = get_labels(y_hat)
```

```python
for i in zip(y_corr, y):
    print(i)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_corr,y))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_corr,y))
```
