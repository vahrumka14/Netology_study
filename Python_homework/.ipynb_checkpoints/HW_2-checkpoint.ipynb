```python
# Задание 1

# Дан список с визитами по городам и странам. Напишите код, который возвращает отфильтрованный список geo_logs, 
# содержащий только визиты из России. Считайте, что список geo_logs легко помещается в оперативной памяти.

geo_logs = [
	{'visit1': ['Москва', 'Россия']},
	{'visit2': ['Дели', 'Индия']},
	{'visit3': ['Владимир', 'Россия']},
	{'visit4': ['Лиссабон', 'Португалия']},
	{'visit5': ['Париж', 'Франция']},
	{'visit6': ['Лиссабон', 'Португалия']},
	{'visit7': ['Тула', 'Россия']},
	{'visit8': ['Тула', 'Россия']},
	{'visit9': ['Курск', 'Россия']},
	{'visit10': ['Архангельск', 'Россия']},
]
```

```python
geo_logs[0]['visit1'][1]  # получение значения России
```

```python
s = []
for i in range(len(geo_logs)):    # проходим по списку
    for j in geo_logs[i].keys():  # проходим по словарю
        if (geo_logs[i][j][1]) == "Россия": # если элемент = Россия
            s.append(geo_logs[i])              # то добавляем в список
print(s)
```

```python
# Задание 3

# Список поисковых запросов. Получить распределение количества слов в них. 
# Т. е. поисковых запросов из одного слова 5%, из двух - 7%, из трех - 3% итд.

queries = [

'смотреть сериалы онлайн',

'новости спорта',

'афиша кино',

'курс доллара',

'сериалы этим летом',

'курс по питону',

'сериалы про спорт',

]
```

```python
l = []
for i in queries:
    s = len(i.split()) # считаем кол-во слов в каждом запросе
    l.append(s)       # создаем список из кол-ва слов в каждом запросе
ll = len(l)         # кол-во запросов          
s = list(set(l))    # кол-во уникальных элементов, т.е. запрос из двух слов, из трех и тд 
for j in range(len(s)): # проходимся по уникальному списку
    print("Кол-во слов в поисковом запросе ={:d}; Доля ={:0.2%}".format (s[j],l.count(s[j])/ll)) # и считаем долю
```

```python
# Таже задача только через list comprehension
l = [len(n.split()) for n in queries] # считаем кол-во слов в каждом запросе
ll = len(l)   # кол-во запросов 
s = list(set(l)) # кол-во уникальных элементов, т.е. запрос из двух слов, из трех и тд 
for j in range(len(s)): # проходимся по уникальному списку
    print("Кол-во слов в поисковом запросе ={:d}; Доля ={:0.2%}".format (s[j],l.count(s[j])/ll)) # и считаем долю
```

```python
# Задание 4 *

# Дана статистика рекламных каналов по объемам продаж. Напишите скрипт, который возвращает название канала 
# с максимальным объемом.

# Т. е. в данном примере скрипт должен возвращать 'yandex'.

stats = {'facebook': 55, 'yandex': 120, 'vk': 115, 'google': 99, 'email': 42, 'ok': 98}
```

```python
s = max(stats.values())    # получаем максимальное значение словаря
for k, v in stats.items(): # проходим по словарю, сравнивая с максимальным значением
    if v == s:
        print(k)
```

```python
# Задание 5 *

# Дан поток логов по количеству просмотренных страниц для каждого пользователя. 
# Список отсортирован по ID пользователя. Вам необходимо написать алгоритм, 
# который считает среднее значение просмотров на пользователя. 
# Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей. 
# Учтите, что весь список stream не помещается в оперативную память, 
# т. е. его нужно обрабатывать поэлементно в цикле.

stream = [

'2018-01-01,user1,3',

'2018-01-07,user1,4',

'2018-03-29,user1,1',

'2018-04-04,user1,13',

'2018-01-05,user2,7',

'2018-06-14,user3,4',

'2018-07-02,user3,10',

'2018-03-21,user4,19',

'2018-03-22,user4,4',

'2018-04-22,user4,8',

'2018-05-03,user4,9',

'2018-05-11,user4,11',

]
```

```python
stats = {}
for line in stream:            # создаем словарь, ключ - user, значение - сумма просмотров по user
    stats.setdefault(line.split(",")[1], 0)
    stats[line.split(",")[1]] += int(line.split(",")[2])
stats
sum(stats.values())/len(stats.keys()) # считаем сумму значений по словарю - это сумма всех просмотров и делим на кол-во уникальных пользователей
```

```python
# Задание 6 *

# Дана статистика рекламных кампаний по дням. 
# Напишите алгоритм, который по паре дата-кампания ищет значение численного столбца. 
# Т. е. для даты '2018-01-01' и 'google' нужно получить число 25. 
# Считайте, что все комбинации дата-кампания уникальны, а список stats легко помещается в оперативной памяти.

stats = [

['2018-01-01', 'google', 25],

['2018-01-01', 'yandex', 65],

['2018-01-01', 'market', 89],

['2018-01-02', 'google', 574],

['2018-01-02', 'yandex', 249],

['2018-01-02', 'market', 994],

['2018-01-03', 'google', 1843],

['2018-01-03', 'yandex', 1327],

['2018-01-03', 'market', 1764],

]
```

```python
key = ['2018-01-01', 'google']
k = [line[2] for line in stats if line[0] == key[0] and line[1] == key[1]]
print(k) 
```

```python
# Бонусный вариант *

# Рассмотрите обобщенный вариант этой задачи: имеется таблица из n столбцов. 
# Первые n-1 столбцов - признаки (комбинации этих признаков в таблице уникальные). 
# Последний столбец - значение. Вам необходимо по набору n-1 признаков найти 
# соответствующую строчку в таблице и вернуть значение из правого столбца. 
# Учтите, что исходная таблица может быть любого размера.

# Одно из применений задачи - объединение двух таблиц по n столбцов, 
# одна из которых помещается в оперативной памяти. Например, 7 столбцов по 1 и 10 миллионов строк.
```

```python
n = len(stats[0])
key = ['2018-01-01', 'google']
for i in stats:
    key_ = ''
    for j in range(n-1):
        key_ = key_ + i[j]
    if key_ == key[0]+key[1]:
        print(i[(n-1)])
```
