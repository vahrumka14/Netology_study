Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_with_movies = ratings.merge(movies, on='movieId', how='left')
ratings_with_movies.head()
```

```python
ratings.shape
```

```python
ratings_with_movies.shape
```

```python
ratings_with_movies[ratings_with_movies['rating'] == 5.0]. \
groupby('title').count().sort_values('rating', ascending = False).head(10)
```

```python
ratings_with_movies[(ratings_with_movies['rating'] == 5.0) & \
                    (ratings_with_movies['title'] == 'Shawshank Redemption, The (1994)')].shape
```

Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power.shape
```

```python
power_filter = power[(power['country'].isin(['Lithuania', 'Latvia', 'Estonia']))& \
      (power['category'].isin([4, 12, 21]))& \
      ((power['year'] >= 2005) & (power['year'] <= 2010))& \
      (power['quantity'] > 0)]
power_filter.head(15)
```

```python
power_filter.quantity.agg('sum')
```
