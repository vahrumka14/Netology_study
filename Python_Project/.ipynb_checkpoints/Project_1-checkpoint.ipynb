4. Визуализировать несколько взаимосвязей между переменными

5. С использованием t-test статистики проверить взаимосвязи между несколькими переменными

```python
sts.ttest_ind(df['number_project'], df['average_montly_hours'])
```

```python
sts.ttest_ind(df['satisfaction_level'], df['average_montly_hours'])
```

У всех гипотез pvalue=0.0, значит, мы не можем говорить о взаимосвязи между показателями.
