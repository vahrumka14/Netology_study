```python
# Интересно также посмотреть, как соотносится тип стекла и конкретный элемент в составе
# В частности, исследуем элемент барий (так как этот элемент имеет наибольшее кол-во значений равных 0)
```

```python
df3= df.loc[df['Ba']==0]
df3 = df3.groupby('Type').count()['Ba'].sort_values(ascending=False)
```

```python
df5 = df.groupby('Type').count()['RI'].sort_values(ascending=False)
```

```python
pd.DataFrame(df5).merge(pd.DataFrame(df3), how = 'left', on = 'Type')
```

```python
# Видно, что присутствие бария хорошо характеризует 7 тип стекла
```

```python
df33= df.loc[df['K']==0]
df33.groupby('Type').count()['K'].sort_values(ascending=False)
```

```python
df333= df.loc[df['Fe']==0]
df333.groupby('Type').count()['Fe'].sort_values(ascending=False)
```

```python
corr_matrix = df.corr()
corr_matrix
```

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.heatmap(corr_matrix, annot = True)
```

```python
# Видно, что на показатель преломления сильно влияет уровень кальция. 
# Есть обратная зависимость между типом стекла и уровнем магния
```

```python
g = sns.pairplot(data=df, hue='Type', palette = 'seismic',
                 height=2,diag_kind = 'kde',diag_kws=dict(shade=True),plot_kws=dict(s=10) )
g.set(xticklabels=[])
```

```python
# Опять видим, что есть зависимость между показателем преломления и уровнем кальция.
```

```python
from sklearn import preprocessing
# нормализуем значения перед отрисовкой Type
min_max_scaler = preprocessing.MinMaxScaler()
column_sels = ['Ca']
x = df.loc[:,column_sels]
y = df['RI']
x = pd.DataFrame(data=min_max_scaler.fit_transform(x), columns=column_sels)
fig, axs = plt.subplots(ncols=1, nrows=1, figsize=(20, 10))
index = 0
for i, k in enumerate(column_sels):
    sns.regplot(y=y, x=x[k])
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```
