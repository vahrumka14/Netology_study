```python
# 6.Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test.
```

```python
price_gas = data.loc[data['fuel-type'] == 'gas', 'price']
price_diesel = data.loc[data['fuel-type'] == 'diesel', 'price']
```

```python
sts.ttest_ind(price_gas, price_diesel)
```

```python
atmo = data[data['aspiration'] == 'std']['city-mpg']
turbo = data[data['aspiration'] == 'turbo']['city-mpg']
```

```python
sts.ttest_ind(atmo, turbo)
# Отвергаем гипотезу из-за pvalue
```

```python
data['city-mpg'].std()*2 # 2 сигмы
```

```python
inside_2sg = data[data['city-mpg'].std()*2<data['city-mpg']]
```

```python
# Отбрасываем выбросы
atmo = inside_2sg[inside_2sg['aspiration'] == 'std']['city-mpg']  
turbo = inside_2sg[inside_2sg['aspiration'] == 'turbo']['city-mpg']
```

```python
sts.ttest_ind(atmo, turbo)
```
