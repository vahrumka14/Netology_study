5. С использованием t-test статистики проверить взаимосвязи между несколькими выборками (двумя или больше)

```python
hr_left = hr[hr.left==1]['satisfaction_level']
hr_not_left = hr[hr.left==0]['satisfaction_level']
```

```python
len(hr_left), len(hr_not_left)
```

```python
hr_left = hr_left.sample(3571)
hr_not_left = hr_not_left.sample(3571)
```

```python
plt.figure(figsize=(10, 7))
plt.hist([hr_left, hr_not_left], bins = 40, label=['hr_left', 'hr_not_left'])
plt.legend(loc='upper right')
plt.show()
```

```python
stats.ttest_ind(hr_left, hr_not_left, equal_var=False)
```

Уровень pvalue<0.05, можем отвергнуть нулевую гипотезу о равенстве средних, значит, что уровень удовлетворенности у уволившихся сотрудников и оставшихся в компании различается. 

```python
hr_left = hr[hr.left==1]['average_montly_hours']
hr_not_left = hr[hr.left==0]['average_montly_hours']
```

```python
len(hr_left), len(hr_not_left)
```

```python
hr_left = hr_left.sample(3571)
hr_not_left = hr_not_left.sample(3571)
```

```python
plt.figure(figsize=(9, 7))
plt.hist([hr_left, hr_not_left], bins = 40, label=['hr_left', 'hr_not_left'])
plt.legend(loc='upper right')
plt.show()
```

```python
stats.ttest_ind(hr_left, hr_not_left, equal_var=False)
```

Уровень pvalue<0.05, можем отвергнуть нулевую гипотезу о равенстве средних, значит, что среднее количество отработанных часов за месяц у уволившихся сотрудников и оставшихся в компании различается.

10. Проверить несколько гипотез для данных, чтобы понять почему сотрудники покидают компанию

Гипотеза 1. О том что на уход из компании влияет степень удовлетворенности.

```python
hr_left = hr[hr.left==1]
```

```python
round(hr_left[hr_left.satisfaction_level<0.5].left.count()/len(hr_left)*100, 1)
```

```python
hr_not_left = hr[hr.left==0]
```

```python
round(hr_not_left[hr_not_left.satisfaction_level<0.5].left.count()/len(hr_not_left)*100, 1)
```

Действительно, среди уволившихся уровень удовлетворенности ниже 50% был у 71%, 
в то время как у сотрудников, которые продолжили работать в компании такой уровень был только у 18%. 

Гипотеза 2. На уход влияет факт повышения в течение 5 лет. 

```python
round(hr_left[hr_left.promotion_last_5years==1].left.count()/len(hr_left)*100, 1)
```

```python
round(hr_not_left[hr_not_left.promotion_last_5years==1].left.count()/len(hr_not_left)*100, 1)
```

Среди уволившихся меньше процент людей, которые получили повышение за последние 5 лет. 

Гипотеза 3. На увольнение влияет уровень зарплаты

```python
hr.groupby(['salary', 'left']).count()['sales']
```

```python
round(hr_left[hr_left.salary == 'high'].left.count()/len(hr_left)*100, 1)
```

```python
round(hr_left[hr_left.salary == 'low'].left.count()/len(hr_left)*100, 1)
```

```python
round(hr_left[hr_left.salary == 'medium'].left.count()/len(hr_left)*100, 1)
```

```python
round(hr_not_left[hr_not_left.salary == 'high'].left.count()/len(hr_not_left)*100, 1)
```

```python
round(hr_not_left[hr_not_left.salary == 'low'].left.count()/len(hr_not_left)*100, 1)
```

```python
round(hr_not_left[hr_not_left.salary == 'medium'].left.count()/len(hr_not_left)*100, 1)
```

У уволившихся процент людей с высокой зарплатой был ниже, чем у людей, которые остались в компании (2.3% и 10.11% соответственно).
Также у уволившихся процент людей с низкой зарплатой был ниже (61% и 45% соответственно).
