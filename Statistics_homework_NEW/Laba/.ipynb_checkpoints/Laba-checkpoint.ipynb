```python
# 6. Выдвинуть несколько гипотез по набору имеющихся переменных и проверить эти гипотезы с использованием t-test.
```

По матрице корреляции у нас цена сильно зависит, имеет прямую корреляцию от engine-size(0.87) и curb-weight(0.84).

Проверим с помощью t-test, что цены на мощные машины отличаются от цен на менее мощные. 

```python
engine_max = df[df['engine-size']>150]['price']
engine_min = df[df['engine-size']<=150]['price']
```

```python
stats.ttest_ind(engine_max, engine_min)
```

Действительно это так, так как pvalue у нас меньше 0.05 и мы не можем говорить о схожести групп.  

```python
curb_weight_max = df[df['curb-weight']>2500]['price']
curb_weight_min = df[df['curb-weight']<=2500]['price']
```

```python
stats.ttest_ind(curb_weight_max, curb_weight_min)
```

Тоже не можем говорить о схожести в группах.
